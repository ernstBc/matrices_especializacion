```python
# lista de los municipios en el dataset 
df['MUNICIPIO'].unique()
```

Se tiene datos de todos los 125 municipios de jalisco.

Los datos marcados como "np.NaN" son los datos agregados para todo el Estado.

```python
# separar los datos de todo el estado de jalisco de sus municipos
jalisco_agg=df[df['MUNICIPIO'].isnull()]

print('Numero de observaciones para el estado de jalisco como agregados:, ',  jalisco_agg.shape[0])
jalisco_agg.head(3)
```

```python
# codificar al estado de kalisco como el municipio 0
jalisco_agg['MUNICIPIO']=jalisco_agg['MUNICIPIO'].fillna(0)
```

```python
muns=df[df['MUNICIPIO'].notnull()]

print('Numero de observaciones en todal para munucipios:, ',  muns.shape[0])
muns.head(3)
```

La variable H001A indica la cantidad de trabjadores empleados en el sector indicado por la variable CODIGO

```python
vars_relevantes=['ENTIDAD', 'MUNICIPIO','CODIGO', 'ID_ESTRATO', 'H001A']
```

```python
# concatenar los dos dataframes de nuevo para exportarlos
df_trabajo=pd.concat([jalisco_agg, muns], axis=0)
df_trabajo=df_trabajo.loc[:, vars_relevantes]
```

```python
# formatear correctamente los datos
df_trabajo['MUNICIPIO']=df_trabajo['MUNICIPIO'].astype(int)
df_trabajo
```

```python
df_trabajo.to_csv(BASE_DATOS_DIR, index=False, mode='w')
```
